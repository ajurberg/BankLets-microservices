```python
import pandas as pd
import numpy as np
import os
```

```python
# Open multiple csv files from a folder and concatenate them into one single dataframe
def concat_csv_files(folder_path):
  """Sorts by name and concatenates multiple csv files from a folder into one single dataframe.

  Args:
    folder_path: The path to the folder containing the csv files.

  Returns:
    A single dataframe containing the data from all the csv files in the folder.
  """
  # Get a list of all the csv files in the folder.
  csv_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.csv')]

  # Read each csv file into a dataframe
  dataframes = [pd.read_csv(file) for file in sorted(csv_files)]

  # Concatenate the dataframes into one single dataframe.
  concat_df = pd.concat(dataframes, ignore_index=True)

  return concat_df
```

```python
# Reads and concatenate multiple csv files
dfs = concat_csv_files(os.getcwd())
```

```python
# Crate a 'Treatment' column
dfs['Treatment'] = dfs['Label'].str.replace('C2-2022_08_09_Adesao.lif-','')
dfs['Treatment'] = dfs['Treatment'].str.replace('[0-9][0-9][0-9]*CL25_Faloidina_A633_adesao_2H_','')
dfs['Treatment'] = dfs['Treatment'].str.replace('[0-9][0-9][0-9]*CL25_Faloidina_A633_adesao_', '')
dfs['Treatment'] = dfs['Treatment'].str.replace('[0-9][0-9][0-9]*4-39_Faloidina_A633_adesao_','')
dfs['Treatment'] = dfs['Treatment'].str.replace('[0-9][0-9][0-9]*7-10_Faloidina_A633_adesao_','')
dfs['Treatment'] = dfs['Treatment'].str.replace('-GDF11','-rGDF11')
dfs['Treatment'] = dfs['Treatment'].str.replace('eries[0-9][0-9][0-9]','')
dfs['Treatment'] = dfs['Treatment'].str.replace('_2H','')
dfs['Treatment'] = dfs['Treatment'].str.replace('_20x','')
```

```python
# Genotype column
dfs['Genotype'] = pd.np.where(dfs['Label'].str.contains('[0-9][0-9][0-9]*CL25_Faloidina_'), 'WT',
                  pd.np.where(dfs['Label'].str.contains('[0-9][0-9][0-9]*4-39_Faloidina_'), 'ALK4-KO',
                  pd.np.where(dfs['Label'].str.contains('[0-9][0-9][0-9]*7-10_Faloidina_'), 'ALK7-KO', 0)))

dfs
```
